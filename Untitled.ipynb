```python
import re
```

```python
cost_series = []
for i in data['cost']:
    k = ''
    c = re.findall(r'\d+',i)
    for j in c:
        k = k+str(j)
    cost_series.append(k)
data['cost'] = cost_series
```

```python
data['Area'].unique()
```

```python
area_series = []
for i in data['Area']:
    k = ''
    try:
        c = re.findall(r'\d+',i)
        for j in c:
            k = k+str(j)
        area_series.append(k)
    except:
        area_series.append(i)
data['Area'] = area_series
```

```python
data.head()
```

```python
data['no.of bed'].unique()
```

```python
data['no.of bed'].replace(['Studio','7+'],['0','8'],inplace=True)
```

```python
data['Bathrooms'].unique()
```

```python
data['Bathrooms'].replace(['7+'],['8'],inplace=True)
```

```python
for i in data.columns:
    try:
        data[i] = pd.to_numeric(data[i])
    except:
        pass
```

```python
data.dtypes
```

```python
data['type'].unique()
```

```python
index = data['cost'].groupby(by=data['type']).median().sort_values().index
values = data['cost'].groupby(by=data['type']).median().sort_values().values
type_map = {}
for i,j in zip(index, range(len(values))):
    type_map.update({i:j})
```

```python
data['type'] = data['type'].map(type_map)
```

```python
data.head()
```

```python
data['Location'].unique()
```

```python
index = data['cost'].groupby(by=data['Location']).median().sort_values().index
values = data['cost'].groupby(by=data['Location']).median().sort_values().values
location_map = {}
for i,j in zip(index, range(len(values))):
    location_map.update({i:j})
```

```python
data['Location'] = data['Location'].map(location_map)
```
